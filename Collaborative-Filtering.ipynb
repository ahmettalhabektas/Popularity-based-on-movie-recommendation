# Finding Similar Movies

## User_Based Collaborative Filtering

```python
import pandas as pd 
```

```python
r_cols = ["user_id","movie_id","rating"]
ratings=pd.read_csv("spark/ml-100k/u.data",sep="\t",names=r_cols,usecols=range(3))
print(ratings.head())
m_cols=["movie_id","title"]
movies=pd.read_csv("spark/ml-100k/u.item",sep="|",names=m_cols,usecols=range(2),encoding="latin-1")
print(movies.head())


```

```python
ratings=pd.merge(movies,ratings)
```

```python
ratings.head()
```

```python
ratings.title.nunique(),ratings.user_id.nunique()
```

```python
moviesRatings=ratings.pivot_table(index=["user_id"],columns=["title"],values="rating")
#,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,,index=user_id,columns=film isimleri,Değerlerse ratinge göre olsun
moviesRatings.head()
#burda yeni bir data frame oluşturduk hangi kullanıcının hangi filme kaç rating verdiğini rahatça görüyoruz
```

```python
starWarsRatings=moviesRatings["Star Wars (1977)"]
starWarsRatings.head()
```

```python
starWarsRatings.value_counts(dropna=False)
```

```python
moviesRatings[["101 Dalmatians (1996)","Star Wars (1977)"]].corr()
```

```python
starWarsRatings.head()
```

```python
moviesRatings.head()
```

```python
moviesRatings.corrwith(starWarsRatings)
```

```python
similarMovies=moviesRatings.corrwith(starWarsRatings)
similarMovies=similarMovies.dropna()
#Boşverileri siliyoruz
similarMovies
```

```python
similarMovies.sort_values(ascending=False)
```

```python
import numpy as np
movieStats=ratings.groupby("title").agg({"rating":[np.size,np.mean]})
movieStats.head(10)
```

# Bazı flimleri yalnızca birkaç insan oylamış işte bu sebepten ötürü en az 100 kişinin görüşüne göre sıralayalım

```python
popularMovies=movieStats[movieStats["rating"]["size"]>=100]
```

```python
popularMovies.sort_values(("rating","mean"),ascending=False)
```

```python
similarMovies.head()
```

```python
popularMovies.head()
```

```python
len(similarMovies),len(popularMovies)
```

```python
df=popularMovies.join(pd.DataFrame(similarMovies,columns=["similarity"]))
df.head(10)
```

```python
len(df)
```

```python
df.sort_values(["similarity"],ascending=False)[:15]
```

```python
df["Rank"]=df["similarity"].rank(ascending=False,method="max")
```
